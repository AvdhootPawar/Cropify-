```python
import pandas as pd
```

```python
# Reading the data 

crop_data_path = r'D:\TY\SEM V\EDI\EDAI\EDAI\datasets\cpdata.csv'
fertilizer_data_path = r'D:\TY\SEM V\EDI\EDAI\EDAI\datasets\Fertilizer.csv'

crop = pd.read_csv(crop_data_path)
fert = pd.read_csv(fertilizer_data_path)
```

```python
# Function for lowering the cases
def change_case(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
fert['Crop'] = fert['Crop'].apply(change_case)
crop['label'] = crop['label'].apply(change_case)
```

```python

#make some changes in Crop Dataset

fert['Crop'] = fert['Crop'].replace('mungbeans','mungbean')
fert['Crop'] = fert['Crop'].replace('lentils(masoordal)','lentil')
fert['Crop'] = fert['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
fert['Crop'] = fert['Crop'].replace('mothbean(matki)','mothbeans')
fert['Crop'] = fert['Crop'].replace('chickpeas(channa)','chickpea')
```

```python
extract_labels = []
for i in crop_names_from_fert:
    if i in crop_names:
        extract_labels.append(i)
```

```python

# using extract labesl on crop to get all the data related to those labels
new_crop = pd.DataFrame(columns = crop.columns)
new_fert = pd.DataFrame(columns = fert.columns)
```

```python
import pandas as pd

# Assuming crop is your original DataFrame
new_crop_list = []  # Create an empty list to hold the DataFrames

for label in extract_labels:
    # Filter the DataFrame for the current label and add it to the list
    filtered_df = crop[crop['label'] == label]
    new_crop_list.append(filtered_df)

# Concatenate all DataFrames in the list into a single DataFrame
new_crop = pd.concat(new_crop_list, ignore_index=True)
```

```python
import pandas as pd

new_fert_list = []  # Create an empty list to hold the DataFrames

for label in extract_labels:
    # Filter the DataFrame for the current label
    filtered_row = fert[fert['Crop'] == label].iloc[0]
    new_fert_list.append(filtered_row)

# Concatenate all filtered rows into a single DataFrame
new_fert = pd.concat(new_fert_list, axis=1).T.reset_index(drop=True)
```
